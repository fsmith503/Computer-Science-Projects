ok, generate a new vector eap_h_halved that halves the count for all h words for eap. For non-h words, it just uses the value that is there. Your eap_h_halved vector should be the same length as eap_vector. 

```
#your code
eap_h_halved = []
for word in sorted_items:
  if word[0][0] == "h":
    eap_h_halved.append(word[1][0]/2)
  else:
    eap_h_halved.append(word[1][0])
```
